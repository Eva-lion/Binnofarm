```python
from report_generator import pretty_timeseria
import pandas as pd
import matplotlib

import sys
sys.path.insert(0, '../')

from google_sheets_parser.google_sheets_parser import GSheetsParser
```

```python
parser = GSheetsParser('https://docs.google.com/spreadsheets/d/1eeOypc11bHbW188ICc17laMf_B73i87AQ9emz2wwvFo/edit#gid=168741310')

df = pretty_timeseria(pd.DataFrame(parser.parse()))
df['j'] = df['click_rate'].rolling(window=13, min_periods=13).median()
df['t'] = df['click_rate'].rolling(window=8, min_periods=8).median()
df['l'] = df['click_rate'].rolling(window=5, min_periods=5).median()
df.plot(y=['click_rate', 'j', 't', 'l'], x='date')
```
